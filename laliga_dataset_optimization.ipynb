
We will create a column of points local in Season (T1PliS) and points visitor in Season (T1PviS) for the local and visitor team.
Each team and season will have a column with the points they have achieved until that round.
It will be normalized (respecting the maximum possible points until that round)

```python
from tqdm import tqdm

for season in tqdm(list_seasons):
    df_laliga.loc[(df_laliga['season'] == season), 'T1PliS'] = 0
    df_laliga.loc[(df_laliga['season'] == season), 'T1PviS'] = 0
    df_laliga.loc[(df_laliga['season'] == season), 'T2PliS'] = 0 #T1 és el local i T2 el visitant
    df_laliga.loc[(df_laliga['season'] == season), 'T2PviS'] = 0
    df_laliga.loc[(df_laliga['season'] == season), 'total_punts_loc'] = 0
    df_laliga.loc[(df_laliga['season'] == season), 'total_punts_vis'] = 0
    #fem llista d'equips d'aquella season:
    ll_equips = df_laliga.loc[(df_laliga['season'] == season), 'localTeam'].unique().tolist()
    #print(ll_equips)
    
    
    for equip in ll_equips:
        counter_points_local = 0
        counter_points_visitor = 0
        counter_partits_local = 1 #evitem error partit 0
        counter_partits_visitor = 1
        entrada_local = False
        entrada_visitor = False
        total_counter_points = 0
        for round in range(1,39):
            #comptem els punts d'aquella ronda de l'equip que estem iterant i els sumem a PliS si és local i a PviS si és visitant
            result_aux = df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round) & 
                                       ((df_laliga['localTeam'] == equip) | (df_laliga['visitorTeam']== equip)), 'result'].iloc[0]
            #result_aux és el resultat de l'equip local
            #calculem si l'equip és local o visitant
            eslocal = len(df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round) & 
                                       (df_laliga['localTeam'] == equip)])
            #eslocal --> 1 si és local, 0 si és visitant
            if eslocal > 1: 
                raise Exception('Error: hi ha més d\'un partit de l\'equip com a local en una mateixa jornada')
            elif eslocal == 1:
                #desem els punts a PliS abans d'actualitzar el counter_points:
                df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round) & 
                                       (df_laliga['localTeam'] == equip), 'T1PliS'] = counter_points_local/(3*counter_partits_local)
                df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round) &
                                        (df_laliga['localTeam'] == equip), 'T1PviS'] = counter_points_visitor/(3*counter_partits_visitor) #no incrementem el counter
                
                df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round) &
                              (df_laliga['localTeam']== equip), 'total_punts_loc'] = total_counter_points
                if result_aux == 2:
                    counter_points_local = counter_points_local + 3
                    total_counter_points = total_counter_points + 3

                elif result_aux == 1:
                    counter_points_local = counter_points_local + 1
                    total_counter_points = total_counter_points + 1

                
                if entrada_local == True: #evitem error partit 0
                    counter_partits_local = counter_partits_local + 1
                entrada_local = True
                
                
            elif eslocal == 0:
                #desem els punts a PviS abans d'actualitzar el counter:
                df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round) & 
                                       (df_laliga['visitorTeam'] == equip), 'T2PviS'] = counter_points_visitor/(3*counter_partits_visitor)
                df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round) &
                                        (df_laliga['visitorTeam'] == equip), 'T2PliS'] = counter_points_local/(3*counter_partits_local) #no incrementem el counter
                
                df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round) &
                              (df_laliga['visitorTeam']== equip), 'total_punts_vis'] = total_counter_points
                if result_aux == 1:
                    counter_points_visitor = counter_points_visitor + 1
                    total_counter_points = total_counter_points + 1

                elif result_aux == 0: #el local ha perdut --> visitant ha guanyat
                    counter_points_visitor = counter_points_visitor + 3
                    total_counter_points = total_counter_points + 3

                if entrada_visitor == True: #evitem error partit 0
                    counter_partits_visitor = counter_partits_visitor + 1
                entrada_visitor = True
                
                
                          
                
```

```python
#We create a new variable that is the month of the season in which the match is played. 1 will be August and 12 will be July.
#For this, we take the date of the match and subtract 7 months from it.

df_laliga['month'] = df_laliga['date'].str[3:5]
df_laliga['month'] = df_laliga['month'].astype(int)
if df_laliga['month'].max() > 12:
    raise Exception('Error: hi ha algun mes que és més gran que 12')
if df_laliga['month'].min() < 1:
    raise Exception('Error: hi ha algun mes que és més petit que 1')



df_laliga['month'] = df_laliga['month'] - 7
df_laliga.loc[df_laliga['month'] < 1, 'month'] = df_laliga['month'] + 12
```

 We can also create new columns that indicate the accumulated goals of the local and visiting team until that match.
We will call it "GoalsliS" and "GoalsviS" respectively. (Goals local in Season and Goals visitor in Season)

```python
for season in tqdm(list_seasons):
    df_laliga.loc[(df_laliga['season'] == season), 'GoalsliS'] = 0
    df_laliga.loc[(df_laliga['season'] == season), 'GoalsviS'] = 0
    #fem llista d'equips d'aquella season:
    ll_equips = df_laliga.loc[(df_laliga['season'] == season), 'localTeam'].unique().tolist()
    #print(ll_equips)
    
    for equip in ll_equips:
        counter_goals = 0
        for round in range(1,39):
            #calculem si l'equip és local o visitant
            eslocal = len(df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round) & 
                                       (df_laliga['localTeam'] == equip)])
            #eslocal --> 1 si és local, 0 si és visitant
                        
            if eslocal > 1: 
                raise Exception('Error: hi ha més d\'un partit de l\'equip com a local en una mateixa jornada')
            elif eslocal == 1:
                
                df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round) & 
                                       (df_laliga['localTeam'] == equip), 'GoalsliS'] = counter_goals
                
                #comptem els gols d'aquella ronda de l'equip que estem iterant i els sumem a GoalsliS si és local i a GoalsviS si és visitant
                goals_aux = df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round) & 
                                       ((df_laliga['localTeam'] == equip)), 'localGoals'].iloc[0]
                
                counter_goals = counter_goals + goals_aux
                
            elif eslocal == 0:
                #desem els punts a PviS abans d'actualitzar el counter:
                df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round) & 
                                       (df_laliga['visitorTeam'] == equip), 'GoalsviS'] = counter_goals
                
                goals_aux = df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round) & 
                                       ((df_laliga['visitorTeam'] == equip)), 'visitorGoals'].iloc[0]
                counter_goals = counter_goals + goals_aux
                
#Funciona perfectament. Triga 2 minuts.
                
```

Finally, we will create a column that tells us the total accumulated points in the last 3 games. In the case that there are no 3 previous games, the missing games will be counted as lost (0 points)


```python
for season in tqdm(list_seasons):
    df_laliga.loc[(df_laliga['season'] == season), 'local_3streak'] = 0
    df_laliga.loc[(df_laliga['season'] == season), 'visitor_3streak'] = 0

    ll_equips = df_laliga.loc[(df_laliga['season'] == season), 'localTeam'].unique().tolist()
    
    for equip in ll_equips:
        
        #the vector league points will have 41 positions initialized to 0
        league_points = np.zeros(41)
        
        #We go to round 4 and save the points of rounds 1, 2 and 3 in league_points.
        for round in range(3,41): #iterem per les 3 inicialitzacions de league_points + les 38 jornades
            result_aux = df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round - 2) & 
                                       ((df_laliga['localTeam'] == equip) | (df_laliga['visitorTeam']== equip)), 'result'].iloc[0]
            
            eslocal = len(df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round - 2) & 
                                       (df_laliga['localTeam'] == equip)])
            if result_aux == 0 and eslocal == 1 or result_aux == 2 and eslocal == 0:
                league_points[round] = 0
            elif result_aux == 1:
                league_points[round] = 1
            elif result_aux == 2 and eslocal == 1 or result_aux == 0 and eslocal == 0:
                league_points[round] = 3
            
            if eslocal == 1: #if it's local, we save the points of the 3 previous rounds in local_3streak
                df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round - 2) & 
                                       (df_laliga['localTeam'] == equip), 'local_3streak'] = league_points[round - 3] + league_points[round - 2] + league_points[round - 1]
            elif eslocal == 0:
                df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round - 2) & 
                                       (df_laliga['visitorTeam'] == equip), 'visitor_3streak'] = league_points[round - 3] + league_points[round - 2] + league_points[round - 1]
            
#We assume the error of those reprogrammed matches.(few matches)
```

```python
#Let's compute now two new variables: local_5streak and visitor_5streak. They will be the sum of the points of the last 5 rounds.
#SAME CODE AS BEFORE BUT WITH 5 INSTEAD OF 3
for season in tqdm(list_seasons):
    df_laliga.loc[(df_laliga['season'] == season), 'local_5streak'] = 0
    df_laliga.loc[(df_laliga['season'] == season), 'visitor_5streak'] = 0

    ll_equips = df_laliga.loc[(df_laliga['season'] == season), 'localTeam'].unique().tolist()
    
    for equip in ll_equips:

        league_points = np.zeros(43)
        
        
        for round in range(5,43): # we iterate for the 5 initializations of league_points + the 38 rounds
            result_aux = df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round - 4) & 
                                       ((df_laliga['localTeam'] == equip) | (df_laliga['visitorTeam']== equip)), 'result'].iloc[0]
            
            eslocal = len(df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round - 4) & 
                                       (df_laliga['localTeam'] == equip)])
            
            if result_aux == 0 and eslocal == 1 or result_aux == 2 and eslocal == 0:
                league_points[round] = 0
            elif result_aux == 1:
                league_points[round] = 1
            elif result_aux == 2 and eslocal == 1 or result_aux == 0 and eslocal == 0:
                league_points[round] = 3
            
            if eslocal == 1: # if it's local, we save the points of the 3 previous rounds in local_3streak
                df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round - 4) & 
                                       (df_laliga['localTeam'] == equip), 'local_5streak'] = league_points[round - 5] + league_points[round - 4] + league_points[round - 3] + league_points[round - 2] + league_points[round - 1]
            elif eslocal == 0:
                df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round - 4) & 
                                       (df_laliga['visitorTeam'] == equip), 'visitor_5streak'] = league_points[round - 5] + league_points[round - 4] + league_points[round - 3] + league_points[round - 2] + league_points[round - 1]
                
```

```python
#We create a variable that says if the team was in first or second division the previous year. 
# In the case of not having information, we will put 1. If they were in first, we will put 1. If they were in second, we will put 0.
df_laliga['last_1ra_loc'] = 4 #random
df_laliga['last_1ra_vis'] = 4 #random

for season in tqdm(list_seasons):

    ll_equips = df_laliga.loc[(df_laliga['season'] == season), 'localTeam'].unique().tolist()
    
    for equip in ll_equips:
                               
        #if the previous season is 86-87, 95-96 or 96-97 we don't have information of the previous season.
        if(season in ['1987-88','1996-97','1997-98']):       
            df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['localTeam'] == equip), 'last_1ra_loc'] = 1
            df_laliga.loc[(df_laliga['season'] == season)  & (df_laliga['visitorTeam'] == equip), 'last_1ra_vis'] = 1

        else:#general case
            
            if df_laliga.loc[(df_laliga['season'] == anterior) & (df_laliga['localTeam'] == equip)].empty:
                estava1ra = 0 #the team was NOT in first division the previous year
            else:
                estava1ra = 1 #the team WAS in first division the previous year
                
            for round in range(1,39):
                
                eslocal = len(df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round) & (df_laliga['localTeam'] == equip)])
                if eslocal == 1: 
                    df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round) & (df_laliga['localTeam'] == equip), 'last_1ra_loc'] = estava1ra
                elif eslocal == 0:
                    df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round)  & (df_laliga['visitorTeam'] == equip), 'last_1ra_vis'] = estava1ra
                
    anterior = season #save for the next iteration
    
            
            
```

```python
#We will do a ranking of the season. We will order total_punts from highest to lowest and assign a number being 1 the highest. 
# In case of a tie we will assign more than one number to the same value.

for season in (list_seasons):
    for round in range (1,39):
        ll_total_punts_loc = df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round), 'total_punts_loc'].tolist()
        ll_total_punts_vis = df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round), 'total_punts_vis'].tolist()

        ll_total_punts = ll_total_punts_loc + ll_total_punts_vis

        set_total_punts = set(ll_total_punts)
        ll_total_punts = list(set_total_punts)
        #convert to float 
        ll_total_punts = [float(i) for i in ll_total_punts]
        ll_total_punts.sort(reverse=True)
        #print(ll_total_punts)  
          

        #We assign the position +1 to each team that has that score.
        df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round), 'Rk_loc'] = -1
        df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round), 'Rk_vis'] = -1
        
        df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round), 'Rk_loc'] = df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round), 'total_punts_loc'].apply(lambda x: list(ll_total_punts).index(x) + 1)
        df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round), 'Rk_vis'] = df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round), 'total_punts_vis'].apply(lambda x: list(ll_total_punts).index(x) + 1)
        

```

```python
#We create a variable that says if the team was in the top 6 the previous year.
dic_classification = {}
for season in (list_seasons):
    #We pick the total_punts_loc and total_punts_vis of round 38 of each season and add the value of result. We will create
    # a dictionary that will save as a key each season and as a value a dictionary with the 20 teams and their 20 positions.
    dic_classification[season] = {}
    ll_equips = df_laliga.loc[(df_laliga['season'] == season), 'localTeam'].unique().tolist()
    for equip in ll_equips:
        eslocal = len(df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == 38) & 
                                       (df_laliga['localTeam'] == equip)])
        
        result_aux = df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == 38) & 
                                       ((df_laliga['localTeam'] == equip) | (df_laliga['visitorTeam']== equip)), 'result'].iloc[0]
        
        if result_aux == 0 and eslocal == 1 or result_aux == 2 and eslocal == 0:
                punts_38 = 0
        elif result_aux == 1:
                punts_38 = 1
        elif result_aux == 2 and eslocal == 1 or result_aux == 0 and eslocal == 0:
                punts_38 = 3
        
        if eslocal == 1:
            loc_38_aux = df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == 38) & (df_laliga['localTeam']== equip), 'total_punts_loc'].iloc[0]
            dic_classification[season][equip] = loc_38_aux + punts_38
        elif eslocal == 0:
            vis_38_aux = df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == 38) & (df_laliga['visitorTeam']== equip), 'total_punts_vis'].iloc[0]  
            dic_classification[season][equip] = vis_38_aux + punts_38
            
#We will do a ranking of the 6 teams with the most points, and we will pass the average.
anterior = None
dic_top6 = {}
for season in (list_seasons):
    dic_top6[season] = []
    if anterior != None:
        
        anterior_keys = list(dic_classification[anterior].keys())
        anterior_values = list(dic_classification[anterior].values())

        #print("anterior keys: ", anterior_keys[0])
        #print("anterior values: ", anterior_values)
        # Sort the values in descending order
        sorted_values = sorted(set(anterior_values), reverse=True)
        # Get the keys corresponding to the 6 biggest different values
        
        count = 0
        for value in sorted_values:
            for i in range(len(anterior_values)):
                if anterior_values[i] == value:
                    dic_top6[season].append(anterior_keys[i])
                    count += 1
                    if count == 6:
                        break
            if count == 6:
                break

        #print(dic_top6[season])
    anterior = season
```

```python
df_laliga['last_top6_loc'] = -1
df_laliga['last_top6_vis'] = -1
for season in (list_seasons):
    ll_equips = df_laliga.loc[(df_laliga['season'] == season), 'localTeam'].unique().tolist()
    for equip in ll_equips:
        if equip in dic_top6[season]: #if the team was in the top 6 the previous year
            for round in range(1,39):
                eslocal = len(df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round) & 
                                       (df_laliga['localTeam'] == equip)])
                if eslocal == 1:
                    df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round) & (df_laliga['localTeam'] == equip), 'last_top6_loc'] = 1
                    
                elif eslocal == 0:
                    df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round) & (df_laliga['visitorTeam'] == equip), 'last_top6_vis'] = 1

        else: #case no Top6
            for round in range(1,39):
                eslocal = len(df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round) &
                                        (df_laliga['localTeam'] == equip)])
                if eslocal == 1:
                    df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round) & (df_laliga['localTeam'] == equip), 'last_top6_loc'] = 0
                    
                elif eslocal == 0:
                    df_laliga.loc[(df_laliga['season'] == season) & (df_laliga['round'] == round) & (df_laliga['visitorTeam'] == equip), 'last_top6_vis'] = 0
            
#(unoptimized code)       
            
```
